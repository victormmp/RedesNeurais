```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import abc
import random
import click
from collections import namedtuple
```

### Criação das Classes

```python
samples = 100
features = 2

class_1 = np.array(np.random.normal(2, 2, [samples, features]))
class_2 = np.array(np.random.normal(6, 2, [samples, features]))

y = np.concatenate(([0 for _ in range(100)], [1 for _ in range(100)]))

plt.scatter(class_1[:,0], class_1[:,1])
plt.scatter(class_2[:,0], class_2[:,1])
```

### Definição dos Modelos de Treinamento

#### Classe abstrata de base para novos modelos

```python
class Model(abc.ABC):
    
    weights = None
    
    def __init__(self, x, y, train_proportion=0.7):
        train_index = np.random.randint(0, x.shape[0], int(train_proportion*x.shape[0]))

        self.x_train = x[train_index]
        self.x_test = x[~train_index]

        self.y_train = y[train_index]
        self.y_test = y[~train_index]
    
    @abc.abstractmethod
    def train(self, epochs=100, eta = 0.1, verbose=True):
        pass
```

#### Perceptron

```python
class Perceptron(Model):
    
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
    
    def train(self, epochs=100, eta=0.1, verbose=True):

        # Add the bias parameter
        x_aug = np.column_stack((np.ones(self.x_train.shape[0]), self.x_train))

        # Initialize Weights
        w = np.random.normal(-0.5, 0.5, (x_aug.shape[1]))
        errors = []

        # Start training
        for epoch in range(epochs):
            if verbose:
                click.echo(f'Epoch {epoch + 1} of {epochs}')

            epoch_error = []
            samples = 
            with click.progressbar(random.sample([num for num in range(x_aug.shape[0])], x_aug.shape[0])) as indexes:
                for i in indexes:
                    y_hat = int(np.dot(x_aug[i, :], w) >= 0)
                    error = self.y_train[i] - y_hat
                    w += eta * error * x_aug[i,:]
                    epoch_error.append(abs(error))

            errors.append(np.mean(epoch_error))
            if verbose:
                click.echo(f'   Error: {np.mean(epoch_error):.3f}')
        self.weights = w

        return w, errors
```

### Treinamento

```python
x = np.concatenate([class_1, class_2])
```

```python
perceptron = Perceptron(x, y)
w, error = perceptron.train(epochs=100, verbose=False)

plt.plot(error); plt.xlabel('epoch'); plt.title('Train Error per Epoch'); plt.ylabel(r'Error (%)')
```

## Utilizando MLP

```python
class MLP(Model):
    def __init__(self, hidden_neurons=5, *args, **kwargs):
        super().__init__(*args, **kwargs)
        
        self.weights = namedtuple('Weights', ['hidden', 'out'])
        self.hidden_neurons = hidden_neurons
    
    def train(self, epochs=100, eta=0.1, verbose=True):
        # Add the bias parameter
        x_aug = np.column_stack((np.ones(self.x_train.shape[0]), self.x_train))

        # Initialize Weights
        w_hidden = np.random.normal(-0.5, 0.5, (x_aug.shape[1], self.hidden_neurons))
        w_out = np.random.normal(-0.5, 0.5, (self.hidden_neurons))
        errors = []
 
        # Start training
        for epoch in range(epochs):
            if verbose:
                click.echo(f'Epoch {epoch + 1} of {epochs}')

            epoch_error = []
            for i in random.sample([num for num in range(x_aug.shape[0])], x_aug.shape[0]):
                ## Feedfoward pipeline
                
                # Remap current sample to the new base
                x_remaped = []
                for neuron in range(w_hidden.shape[1]):
                    x_remaped.append(np.dot(x_aug[i, :], w_hidden[:, neuron]))
                x_remaped = np.array(x_remaped)
                
                # Calculate the output
                y_hat = np.tanh(np.dot(x_remaped, w_out) >= 0)
                error = self.y_train[i] - y_hat
                epoch_error.append(abs(error))
    
                ## Backpropagation pipeline
                for w in w_out.shape[0]:
                    w_out[]

            errors.append(np.mean(epoch_error))
            if verbose:
                click.echo(f'   Error: {np.mean(epoch_error):.3f}')
            self.weights.hidden = w_hidden
            self.weights.out = w_out

        return self.weights, errors
```
